# Brent diesel/gasoline relationship in Central Europe

This notebook shows how to analyse basic relationship between diesel and gasoline with Brent in Central Europe. Basic OLS is used for this purpose. Stationarity, autocorrelation, lagged values and so on are not taken into account, as those issues will be analysed in another notebooks in this repository.

Used data are from widely available sources.

```python
from matplotlib import pyplot
import pandas as pd
import numpy as np
import statsmodels.api as sm
from pandas import read_excel
```

```python
df = read_excel('final_oils_python2.xlsx', sheet_name ='LR commodities')
df.set_index(df['Date'],inplace=True)
df.index = pd.to_datetime(df.index).date

del df['Date']
```

```python
commodity_list=['Diesel Czech Republic', 'Diesel Hungary', 'Diesel Slovakia','Diesel Poland']
colors=['darkblue','green', 'blue', 'red']

df['const']=1

for i in range(len(commodity_list)):

    reg=sm.OLS(endog=df[commodity_list[i]],exog=df[['const','Brent']])
    
    type(reg)
    results=reg.fit()
    
    x=results.fittedvalues
    
    x.plot(title=f"{commodity_list[i]}   R2: {round(results.rsquared,2)}",figsize=(12,4),
           alpha=0.5,color=colors[i],linestyle="--",label="OLS fitted values",legend=True)
    
    df[commodity_list[i]].plot(color=colors[i],label="Original values",legend=True)
    pyplot.show()
```

```python
commodity_list=['Gasoline Czech Republic', 'Gasoline Hungary', 'Gasoline Slovakia','Gasoline Poland']
colors=['darkblue','green', 'blue', 'red']

df['const']=1

for i in range(len(commodity_list)):

    reg=sm.OLS(endog=df[commodity_list[i]],exog=df[['const','Brent']])
    
    type(reg)
    results=reg.fit()
    
    x=results.fittedvalues
    
    x.plot(title=f"{commodity_list[i]}   R2: {round(results.rsquared,2)}",figsize=(12,4),
           alpha=0.5,color=colors[i],linestyle="--",label="OLS fitted values",legend=True)
    
    df[commodity_list[i]].plot(color=colors[i],label="Original values",legend=True)
    pyplot.show()
```
